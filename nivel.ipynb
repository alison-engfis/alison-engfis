# NÍVEL DO GUAÍBA EM UM ANO

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dados.csv')

df.head(21)
```

```python
df['date'] = pd.to_datetime(df['date'], format= '%d/%m/%Y %H:%M:%S')
```

```python
df.head(10)
```

```python
valor_max = 6

# Filtrar valores aceitáveis
df_limpo = df[df['water level (meters)'] <= valor_max]
```

```python
df.describe()
```

```python
# Filtrar apenas as medições realizadas às 00:00:00
df_midnight = df[df['date'].dt.hour == 0]

# Plotar gráfico
plt.figure(figsize= (10, 6))
sns.lineplot(data= df_midnight, x= 'date', y= 'water_level_meters', marker= 'o')
plt.title('Nível do Rio Guaíba às 00:00:00')
plt.xlabel('Data')
plt.ylabel('Nível da Água (metros)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
