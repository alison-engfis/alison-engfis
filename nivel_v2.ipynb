# NÍVEL DO GUAÍBA AO LONGO DO TEMPO

```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sph4_out.csv', sep= ',')

df.tail(10)
```

```python
print(df.columns)
```

```python
df.rename(columns= {'% year' : 'year',' month' : 'month', ' day' : 'day', ' hour' : 'hour', ' minute' : 'minute', ' second (GMT/UTC)' : 'second',
                    ' water level (meters)' : 'water_level(m)'}, inplace= True)

df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour', 'minute', 'second']])

df.tail(10)
```

# CRIANDO O GRÁFICO COM MATPLOTLIB COMO FOI PROPOSTO

```python
plt.figure(figsize= (14, 6))
plt.plot(df['datetime'], df['water_level(m)'], linestyle= '-', color= 'b')
plt.xlabel('Data', fontsize= 14)
plt.ylabel('Nível da Água (metros)', fontsize= 14)
plt.title('Nível do Guaíba ao Longo do Tempo', fontsize= 16)
plt.xticks(rotation= 45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
